## 1. **Import Library**
### Untuk menggunakan library pengolahan citra pada python, bisa menggunakan scikit-image dan matplotlib. Gunakan syntax berikut untuk mengimport  library tersebut:

```python
from skimage import color, io
from skimage.transform import rescale
import matplotlib.pyplot as plt
```

## **2. Membaca Citra**
### Download gambar yang telah disediakan di SCeLE kemudian ekstrak dan letakkan seluruh gambar pada direktori tempat anda bekerja.
### Lakukan pembacaan fasilkom.png dan fasilkomgelap.png. Untuk membaca gambar, gunakan perintah io.imread(‘file_name’)​, contoh:

```python
i1 = io.imread('fasilkom.png')
```

## **3. Menampilkan Citra**
### Untuk menampilkan gambar, gunakan fungsi plt.imshow(var_name)​, contoh:

```python
plt.imshow(i1, cmap='gray', vmin=0, vmax=255)
plt.show()
```

## **4. Mengecek Dimensi Citra**

```python
dimension = i1.shape
dimension
```

## **5. Menyimpan Gambar**
### Untuk menyimpan gambar, gunakan fungsi io.imsave(’file_name’, var_name)​, Contoh:

```python
io.imsave('fasilkombaru.png', i1)
```

## **7. Resolusi**
### Resolusi adalah kerapatan pixel pada suatu gambar, semakin besar resolusi spasial maka semakin banyak pixel yang digunakan dalam menampilkan suatu citra.

```python
i3 = rescale(i1, 0.5, anti_aliasing=True)
i4 = rescale(i1, 0.1, anti_aliasing=True)

plt.imshow(i3, cmap='gray')
plt.show()
plt.imshow(i4, cmap='gray')
plt.show()
```

## **8. Konversi Citra**
### Untuk mengkonversi citra, dapat menggunakan beberapa fungsi yang sudah disediakan oleh scikit-image, yaitu sebagai berikut.

| Fungsi | Kegunaan | Format |
| ------ | -------- | ------ |
| rgb2hsv | RGB ke HSV. | y=color.rgb2hsv(x) |
| hsv2rgb | HSV ke RGB. | y=color.hsv2rgb(x) |
| rgb2gray | RGB ke grayscale. | y=color.rgb2gray(x) |
| gray2rgb | Membuat representasi RGB dari image grayscale. | y=color.gray2rgb(x[, alpha]) |



# Selamat Mencoba!

```python
# Normal
i5 = io.imread('cats.jpg')
plt.imshow(i5)
plt.show()

# HSV
i6 = color.rgb2hsv(i5)
plt.imshow(i6)
plt.show()

# Grayscale
i7 = color.rgb2gray(i5)
plt.imshow(i7, cmap="gray")
plt.show()

# Hue Channel
i8 = i6[:, :, 0]
plt.imshow(i8, cmap="hsv")
plt.show()

# Value Channel
i9 = i5[:, :, 2]
plt.imshow(i9, cmap="gray")
plt.show()
```
