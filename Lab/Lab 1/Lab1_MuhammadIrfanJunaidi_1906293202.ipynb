```python
from skimage import io, color, util, exposure, filters, morphology
import numpy as np
import matplotlib.pyplot as plt
```

## **1. [40] Diberikan citra “flower.jpg” yang berukuran 563 x 1000.**
<img src="Citra Tugas Lab 1\sunflower.jpg">

## a. [5] Tampilkan histogram citra tersebut dengan range intensitas 0 - 255

```python
i1 = io.imread("Citra Tugas Lab 1\sunflower.jpg")

gray = util.img_as_ubyte(color.rgb2gray(i1))

plt.title("Original")
plt.hist(gray.flatten(), 256, range=(0,256))
plt.show()
```

## b. [10] Lakukan histogram equalization terhadap citra tersebut kemudian tampilkan histogramnya.

```python
# Equalization
eq_i1 = exposure.equalize_hist(i1)

gray_eq = util.img_as_ubyte(color.rgb2gray(eq_i1))

plt.title("Histogram Equalization")
plt.hist(gray_eq.flatten(), 256, range=(0,256))
plt.show()
```

## c. [10] Lakukan contrast stretching terhadap citra asli kemudian tampilkan histogramnya.

```python
# Contrast Stretching
mn = min(i1.flatten())
mx = max(i1.flatten())
b = int(np.floor(255 / (mx - mn)))
i1_cs = (i1 - mn) * b

gray_cs = util.img_as_ubyte(color.rgb2gray(i1_cs))

plt.title("Contrast Stretching")
plt.hist(gray_cs.flatten(), 256, range=(0,256))
plt.show()
```

## d. [15] Coba ceritakan perbedaan apa yang dapat anda temukan antara metode histogram equalization dan contrast stretching.

### **Histogram Equalization** mengubah pemetaan gray level menjadi lebih menyebar dalam range 0-255. Dapat dilihat bahwa pada hasil **Histogram Equalization** level kontras yang jumlahnya banyak dibuat renggang dan level kontras yang jumlahnya sedikit dibuat rapat. Sedangkan, **Contrast Stretching** memperluas range intensitas citra yang tadinya terbatas menjadi range intensitas penuh.

## **2. [30] Diberikan citra “iris.jpg” berukuran 254 x 199.**

## a. [10] Gunakan Sobel spatial filter dan Prewitt spatial filter untuk menemukan dan menampilkan horizontal edge dari citra tersebut

```python
i2 = io.imread("Citra Tugas Lab 1\iris.jpg")
plt.figure(figsize=(10, 10))

gray2 = color.rgb2gray(i2)

gray2_h_pre = util.img_as_ubyte(filters.prewitt_h(gray2))
gray2_h_sob = util.img_as_ubyte(filters.sobel_h(gray2))

plt.subplot(1,2,1); plt.imshow(gray2_h_pre, cmap="gray")
plt.title('Prewitt'); plt.axis("off")
plt.subplot(1,2,2); plt.imshow(gray2_h_sob, cmap="gray")
plt.title('Sobel'); plt.axis("off")

plt.show()
```

## b. [10] Gunakan Sobel spatial filter dan Prewitt spatial filter untuk menemukan dan menampilkan vertical edge dari citra tersebut

```python
plt.figure(figsize=(10, 10))

gray2_v_pre = util.img_as_ubyte(filters.prewitt_v(gray2))
gray2_v_sob = util.img_as_ubyte(filters.sobel_v(gray2))

plt.subplot(1,2,1); plt.imshow(gray2_v_pre, cmap="gray")
plt.title('Prewitt'); plt.axis("off")
plt.subplot(1,2,2); plt.imshow(gray2_v_sob, cmap="gray")
plt.title('Sobel'); plt.axis("off")

plt.show()
```

## c. [10] Berdasarkan hasil 2a. dan 2b. temukan dan tampilkan edge keseluruhan dari citra tersebut

```python
plt.figure(figsize=(10, 10))

gray2_pre = util.img_as_ubyte(filters.prewitt(gray2))
gray2_sob = util.img_as_ubyte(filters.sobel(gray2))

plt.subplot(1,2,1); plt.imshow(gray2_pre, cmap="gray")
plt.title('Prewitt'); plt.axis("off")
plt.subplot(1,2,2); plt.imshow(gray2_sob, cmap="gray")
plt.title('Sobel'); plt.axis("off")

plt.show()
```

## **3. [30] Diberikan gambar “starfish.png” berukuran 256 x 384 sebagai berikut:**
<img src="Citra Tugas Lab 1\starfish.png">

## a. [7.5] Terapkan median filter pada citra tersebut dengan ukuran kernel 9x9


```python
i3 = io.imread("Citra Tugas Lab 1\starfish.png")

gray3 = color.rgb2gray(i3)

gray3_med_fi = filters.rank.median(gray3, selem=morphology.square(9))

plt.title("Median Filter")
plt.imshow(gray3_med_fi, cmap="gray"); plt.axis("off")
plt.show()
```

## b. [7.5] Terapkan mean filter pada citra tersebut dengan ukuran kernel 9x9


```python
gray3 = color.rgb2gray(i3)

gray3_mea_fi = filters.rank.mean(gray3, selem=morphology.square(9))

plt.title("Mean Filter")
plt.imshow(gray3_mea_fi, cmap="gray"); plt.axis("off")
plt.show()
```

## c. [15] Ceritakan perbedaan citra yang dihasilkan pada soal 3a. dan 3b. Filter manakah yang hasilnya lebih baik menurut Anda? Berikan pula alasannya.

### Citra yang dihasilkan dengan **Median Filter** maupun **Mean Filter** membuat citra menjadi buram. Filter yang hasilnya lebih baik adalah **Median Filter** karena selain gambar diburamkan, noise yang sebelumnya ada menjadi hilang. Sedangkan, pada gambar yang dilakukan **Mean Filter** memburamkan gambar tetapi noisenya masih ada. Hal ini dikarenakan **Mean Filter** terpengaruh oleh noise pada gambar pada saat melakukan _blurring_ di mana **Mean Filter**, sesuai dengan namanya, **Mean Filter** mengambil rata-rata piksel pada area tertentu. **Median Filter** tidak terpengaruh oleh noise karena piksel yang nilainya sangat tinggi maupun sangat rendah tidak akan terpengaruh nilai tengah.
